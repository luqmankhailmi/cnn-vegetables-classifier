```python
# Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import layers
```

```python
# Define Dataset Paths (Data Collection)
data_train_path = 'C:/Users/nnbaj/Desktop/CSC583/Vegetable_classifier/Vegetable/train'
data_val_path = 'C:/Users/nnbaj/Desktop/CSC583/Vegetable_classifier/Vegetable/validate'
data_test_path = 'C:/Users/nnbaj/Desktop/CSC583/Vegetable_classifier/Vegetable/test'
```

```python
# Set Image Fixed Size (Data Preprocessing)
img_width = 180
img_height =180 
```

```python
# Load Dataset (Data Collection + Preprocessing)
data_train = tf.keras.utils.image_dataset_from_directory(
    data_train_path,
    shuffle=True,
    image_size=(img_width, img_height),
    batch_size=8,
    validation_split=False)
```

```python
# Extract Class Name ( Label Mapping)
# Purpose: Automatically detects folder names as class labels 
data_cat = data_train.class_names
```

```python
data_cat
```

```python
# Load Dataset (Data Collection + Preprocessing)
data_val = tf.keras.utils.image_dataset_from_directory(data_val_path,
                                                       image_size=(img_height,img_width),
                                                       batch_size=8,
                                                        shuffle=False,
                                                       validation_split=False)
```

```python
# Load Dataset (Data Collection + Preprocessing)
data_test = tf.keras.utils.image_dataset_from_directory(
data_test_path,
    image_size=(img_height,img_width),
    shuffle=False,
    batch_size=8,
    validation_split=False
)
```

```python
# Print images from the training dataset  

image, labels = next(iter(data_train))

plt.figure(figsize=(10,10))

plt.figure(figsize=(10,10))
for i in range(min(9, len(image))):
        plt.subplot(3, 3, i+1)
        plt.imshow(image[i].numpy().astype("uint8"))
        label_index = labels[i].numpy()
        plt.title(data_cat[label_index])
        plt.axis("off")
        
plt.tight_layout()
plt.show()
```

```python
from tensorflow.keras.models import Sequential
```

```python
# Print data scope
data_train
```

```python
# Model Building
model = Sequential([
    layers.Rescaling(1./255),
    layers.Conv2D(16, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(32,3, padding='same',activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(64, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    layers.Flatten(),
    layers.Dropout(0.2),
    layers.Dense(128),
    layers.Dense(len(data_cat))
                  
])
```

```python
model.compile(optimizer='adam', loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True), metrics=['accuracy'])
```

```python
from PIL import Image
import os

def remove_invalid_images(folder_path):
    deleted = 0
    for root, _, files in os.walk(folder_path):
        for file in files:
            filepath = os.path.join(root, file)
            try:
                with Image.open(filepath) as img:
                    img.verify()  # Verify it is a valid image
            except Exception as e:
                print(f"❌ Deleting invalid file: {filepath}")
                os.remove(filepath)
                deleted += 1
    print(f"\n✅ Clean-up complete. {deleted} invalid files deleted.")

# Your dataset root path here:
remove_invalid_images('C:/Users/nnbaj/Desktop/CSC583/Vegetable_classifier/Vegetable')
```

```python
import os

extensions = set()

for root, _, files in os.walk('C:/Users/nnbaj/Desktop/CSC583/Vegetable_classifier/Vegetable'):
    for file in files:
        ext = os.path.splitext(file)[-1].lower()
        extensions.add(ext)

print("File extensions found in dataset:", extensions)

from PIL import Image
import os

dataset_path = 'C:/Users/nnbaj/Desktop/CSC583/Vegetable_classifier/Vegetable'
bad_files = []

for root, _, files in os.walk(dataset_path):
    for file in files:
        filepath = os.path.join(root, file)
        try:
            with Image.open(filepath) as img:
                img.verify()  # Check if the image can be opened
        except Exception as e:
            print(f"❌ Corrupted or unreadable: {filepath}")
            bad_files.append(filepath)

# Delete the bad ones
for path in bad_files:
    os.remove(path)
    print(f"🗑️ Deleted: {path}")

print(f"\n✅ Done. {len(bad_files)} corrupted files removed.")

import tensorflow as tf
import os

broken_files = []
dataset_path = 'C:/Users/nnbaj/Desktop/CSC583/Vegetable_classifier/Vegetable'

for root, _, files in os.walk(dataset_path):
    for file in files:
        file_path = os.path.join(root, file)
        try:
            img_raw = tf.io.read_file(file_path)
            _ = tf.image.decode_image(img_raw)  # Try decoding like TensorFlow does
        except Exception as e:
            print(f"❌ TensorFlow can't read: {file_path}")
            broken_files.append(file_path)

print(f"\n✅ Scan complete. Found {len(broken_files)} broken image(s).")

# Optional: Delete them
for f in broken_files:
    os.remove(f)
    print(f"🗑️ Deleted: {f}")

```

```python
epochs_size = 3
history = model.fit(data_train, validation_data=data_val, epochs=epochs_size)
```

```python
epochs_range = range(epochs_size)
x = np.arange(len(epochs_range))  # positions for the bars
width = 0.35  # width of the bars

plt.figure(figsize=(10, 6))

# 📊 Accuracy Bar Chart
plt.subplot(1, 2, 1)
plt.bar(x - width/2, history.history['accuracy'], width, label='Training Accuracy', color='skyblue')
plt.bar(x + width/2, history.history['val_accuracy'], width, label='Validation Accuracy', color='orange')
plt.title('Accuracy per Epoch')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.xticks(x, [f"Epoch {i+1}" for i in x])
plt.ylim(0, 1)  # Accuracy values range from 0 to 1
plt.legend()

# 📉 Loss Bar Chart
plt.subplot(1, 2, 2)
plt.bar(x - width/2, history.history['loss'], width, label='Training Loss', color='lightgreen')
plt.bar(x + width/2, history.history['val_loss'], width, label='Validation Loss', color='salmon')
plt.title('Loss per Epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.xticks(x, [f"Epoch {i+1}" for i in x])
plt.legend()

plt.tight_layout()
plt.show()
```
